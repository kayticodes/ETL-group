```python
import pymongo
import pandas as pd
import json
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.music
music_data = db.music_data
```

```python
music_df = pd.read_csv("music_data_final.csv")
music_df.head()
```

```python
music_df_track = music_df[["Track Name","Artist","album","Genre"]].drop_duplicates()
```

```python
music_df_track.head()
```

```python
for index , row in music_df_track.iterrows():
    mongoDoc={}
    mongoDoc["Track"] = row["Track Name"]
    mongoDoc["Artist"] = row["Artist"]
    mongoDoc["Album"]  = row["album"]
    mongoDoc["Genre"]  = row["Genre"] 
    mongoDoc["Statistics"] = json.loads( music_df.loc[music_df["Track Name"]==row["Track Name"]][["Date","Position","Streams","Country"]].T.to_json())
    music_data.insert_one(mongoDoc)
```

```python
collection = music_data.find()
for row in collection:
    print(row)
```
